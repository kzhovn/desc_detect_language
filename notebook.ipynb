```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from langdetect import detect
import polyglot
import pycountry
import unicodedata
import pdb
```

```python
df = pd.read_csv('group_descriptions.csv', error_bad_lines = False)
```

```python
df = df.dropna(subset = ['description'])
df.description = df['description'].map(lambda text: BeautifulSoup(text, 'lxml').get_text())

pd.set_option('display.max_colwidth', 1000)
df.head()
```

```python
df.description.replace('', np.nan, inplace=True)
df.description.replace(' ', np.nan, inplace=True)

df = df.dropna(subset = ['description'])
df.info()
```

```python
df['language'] = ""
df.head()
```

```python
def detect_error(str):
    try:
        return detect(str)
    except:
        return "Error"

sample_df = df.sample(n = 1000, random_state = 1)
sample_df['language'] = sample_df['description'].apply(detect_error)
```

```python
from polyglot.detect import Detector

def detect_error_polyglot(str):
    try:
        return Detector(str).language
    except:
        return "Error"

sample_df2 = df.sample(n = 1000, random_state = 1)
sample_df2['language'] = sample_df2['description'].apply(detect_error_polyglot)
```
